```python
def plot_top_patients(df, n=10):
    df["Visit (Antenatal Care Followup)"].value_counts().head(n).sort_values().plot(kind="barh")
    plt.xlabel("Number of Records")
    plt.title(f"Top {n} Patients by Record Count")
    plt.tight_layout()
    plt.show()

plot_top_patients(df, 10)
```
